## Generalize solution

```python
def generalized_separation(image_path, box, iterations, plot=True):
    img_orig = cv2.imread(image_path)
    img = img_orig.copy()
    mask = np.zeros(img.shape[:2],np.uint8)
    bgdModel = np.zeros((1,65), np.float64)
    fgdModel = np.zeros((1,65), np.float64)
    cv2.grabCut(img, mask, box, bgdModel, fgdModel, iterations, cv2.GC_INIT_WITH_RECT)
    mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
    img = img * mask2[:,:,np.newaxis]
    
    
    if plot:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_orig = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)
        f = plt.figure(figsize=(20, 10))
        f.add_subplot(1,2, 1)
        plt.imshow(img)
        
        f.add_subplot(1,2, 2)
        plt.imshow(img_orig)
        plt.show(block=True)
        
```

## img1

```python
img_path = './images3/obj1.jpg'
box = (35, 30, 425, 220)

generalized_separation(img_path, box=box, iterations=30)
```
